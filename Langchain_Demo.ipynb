## Simple AI Assistant

```python
while True:
    question=input("Type Your Question.If you want to quit the chat write quite")
    if question !="quit":
        print(llm_1.invoke(question).content)
    else:
        print("Goodbye!")
        break
```

```python
from langchain_core.chat_history import BaseChatMessageHistory
from langchain_core.chat_history import InMemoryChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_core.messages import AIMessage
```

```python
store={}
```

```python
def get_session_history(session_id:str)->BaseChatMessageHistory:
    if session_id not in store:
        store[session_id]=InMemoryChatMessageHistory()
    return store[session_id]
```

```python
config={"configurable":{"session_id":"firstchat"}}
```

```python
model_with_memory=RunnableWithMessageHistory(llm,get_session_history)
```

```python
model_with_memory.invoke(("Hi! I am ashutosh singh"),config=config).content
```

```python
model_with_memory.invoke(("Hi!,tell what is my name"),config=config).content
```

```python
from langchain.agents import AgentType
from langchain.agents import load_tools
from langchain.agents import initialize_agent
```
